Читаем файл библиотекой Pandas, с кодировкой - cp1251 и разделителем -";"

```python
df=pd.read_csv("obespechennost_vrachami_na_10000_chelovek_chelovek_permstat_7_dekabrya_2013_0.csv",encoding="cp1251",delimiter=";")
df.head()
```

Получите максимальные и минимальные значения для датафреймов и для каждого  столбца где это возможно


```python
df.min(), df.max()
```

```python
df['Значение'] = df['Значение'].str.replace(',','.')
```

```python
df['Значение']=df['Значение'].astype(float)
```

Получите данные с данными в столбце 'Значение' более 70 и сохраните его в новый фрейм dff 


```python
dff = df[df['Значение']>70]
dff.sort_values('ОКАТО')
```

```python
set(dff['ОКАТО'])
```

Проверьте сколько значений лежит в диапазоне от 89 до 100


```python
len(df[df["Значение"].between(89, 100)])
```

Проверьте сколько значений равно 70


```python
len(dff[dff['Значение']==70])
```

Проверьте  сколько значений не равно 90


```python
len(dff[dff['Значение']!=90])
```

Составьте Лямбда функцию, проверяющее это условие (неравенство 90) и примените ее


```python
dff[dff['Значение'].apply(lambda value: value != 90)]
```

Обработайте данные по паре городов (районов) (по вариантам)


```python
df['ОКАТО'].unique()
```

'Пермский край', 'Гайнский район'


```python
df_reg= df[(df['ОКАТО']=='Пермский край') | (df['ОКАТО']=='Гайнский район')]
df_reg.head()
```

Найдите столбец с датой, преобразуйте ее в формат datetime


```python
df_reg['Дата']=pd.to_datetime(df_reg['Дата'])
print(df_reg['Дата'].dtype)
```

Найдите лишний столбец с текстом 'Обеспеченность врачами на 10000 
человек'  и удалите его


```python
df_reg.drop('Статистические показатели',axis=1,inplace=True)
df_reg
```

 Удалите столбец 'ОКАТО'


```python
df_reg.drop('ОКАТО',axis=1)
```

 Задайте числовой целый формат столбцу 'Значение'



```python
df_reg['Значение'] = df_reg['Значение'].astype(int)
df_reg.head()
```

Отсортируйте по городам (районам)



```python
df_reg.sort_values('ОКАТО')
```

Подсчитайте по городам (регионам) среднее и медианное значение, при необходимости округлите до целого методом round() 


```python
round(df_reg.groupby('ОКАТО').mean())
```

```python
df_reg.groupby('ОКАТО').median()
```

Выведите графическим способом полученные результаты по каждому городу в отдельности


```python
df_reg_perm = df_reg[df_reg['ОКАТО']=='Пермский край']
plt.title('Пермский край')
plt.xlabel('Дата')
plt.ylabel('Окато')
plt.plot(df_reg_perm.Дата,df_reg_perm.Значение,'ro')
```

```python
df_reg_gayn = df_reg[df_reg['ОКАТО']=='Гайнский район']
plt.title('Гайнский район')
plt.xlabel('Дата')
plt.ylabel('Окато')
plt.plot(df_reg_gayn.Дата,df_reg_gayn.Значение,'ro')
```
